```python
import pandas as pd
import math
import csv
```

```python
#vote clean up 
file_path = "../Data/Input_Step3/vote_names.csv"
data_df = pd.read_csv(file_path)
vote_set ={"za","fa","ja","nein","beurl","krank","entsch","stein","2a","3«","3a","3»"
#,"3st"}
data_list = data_df.values.tolist()

for data_row in data_list:
    for data in data_row:
        #print(data)
        if isinstance(data, float) and math.isnan(data):
            continue
        element = data.split()
        #print("START",element)
        for eachElement in element:
            #print("-"+eachElement.strip()+"-")
            eleProcess = eachElement.strip().lower().replace(".","")
            if len(eleProcess)<=1:
                continue
            vote_set.add(eleProcess)
print("done")
for vote in vote_set:
    print(vote)
```
