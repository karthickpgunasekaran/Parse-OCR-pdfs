```python
###############################################################################################
0) Remove all the multiple dots
1) Split everything by spaces first
2) Process each word in the split string 
   2a) if there are only two characters or less then append it to next word
   2b) if there are brackets starting and ending add it to prev word
   2c) Check if it belongs to one of the words in vote_types
   2d) If 'Yes' add it to the csv
   2e) If 'No' then check the length of the characters
       2f) if less than 2 then remove it
3) If the total number is not as expected write it to a log file on words which needs to be checked.
4) 
##############################################################################################
```

```python
#Initialize all the libraries needed here
from tabula import read_pdf
from tabula import convert_into
import pandas as pd 
import math
import re
from fuzzywuzzy import fuzz
import logging
logging.basicConfig(level = logging.INFO, filename = "../my.log")
class DataProcessor:
    def __init__(self,data,filename,party_data=None,vote_data=None,exp_cols=None):
        self.data = data
        
        self.party_set = set()
        self.output_filename = filename 
        self.party_list = party_data
        if party_data!=None:
            self.loadPartyNames()
            self.party_cnt =0
        # define vote types here
        if vote_data!=None:
            self.vote_types = vote_data
        else:
            self.vote_types ={"za","fa","ja","nein","beurl","krank","entsch","stein","2a","3«","3a","3»"
,"3st"}
        self.allRows = [[]]
        self.row = []
        self.prev_complete_name = False
        self.missing_rows = 0
        self.total_rows = 0
        if exp_cols!=None:
            self.expected_columns =exp_cols
        self.previous_party = "" 
        
        #self.logger = logging#.getLogger(__name__)

    def checkDashes(self,list_ele):
        new_li = []
        for i in range(0,len(list_ele)):
            if list_ele[i]!="-":
                new_li.append(list_ele[i])
        return new_li
    #process each word to obtain the final structure    
    def addStructure(self,word):
        word_pro =re.sub('[^A-Za-z0-9«»”]+', '',word).lower()
        word_all_pro =re.sub('[^A-Za-z0-9]+', '',word).lower()
        
        #print("processed word : ",word_pro)
        if word_pro in self.vote_types:
            #check for the type of vote
            self.prev_complete_name = False
            self.row.append(word_pro)
        elif word_all_pro in self.vote_types:
            #check for the type of vote
            self.prev_complete_name = False
            self.row.append(word_all_pro)    
        elif word.strip()=="—" or word.strip()=="-":
            #check for the dash
            self.prev_complete_name = False
            self.row.append("-")
            
        elif len(word_all_pro)<=1:
            self.prev_complete_name = False
            #check if its an empty string after removing all the symbols
            return
        
        else:
            #This case is for the names
            two_dots = re.findall("\.{2,}",word)
            word_dot = re.sub('\.{2,}|\-{2,}','',word)
            #print(word+" new: "+word_dot)
            
            #    complete_name = True
            if len(self.row) >1 or self.prev_complete_name:
                self.total_rows =self.total_rows+1
                if len(self.row)==self.expected_columns:
                    if self.previous_party!="":
                        self.row.insert(0,self.previous_party)
                    else:
                        self.row.insert(0,"-")
                    msg = "Added rows: " + ",".join(self.row)      # fully build the message
                    #logger.info(msg)  
                    #logging.info(msg)
                    self.allRows.append(self.row)
                else:
                    new_row = self.checkDashes(self.row)
                    if len(new_row)==self.expected_columns:
                        if self.previous_party!="":
                            new_row.insert(0,self.previous_party)
                        else:
                            new_row.insert(0,"-")
                        self.allRows.append(new_row)
                        msg = " - Modified rows: " + ",".join(new_row)      # fully build the message
                        #logger.info(msg)  
                        #logging.info(msg)
                        #print("ignorned: ",self.row)
                    else:
                        msg = "Ignored rows: " + ",".join(self.row)      # fully build the message
                        #logger.info(msg)  
                        #logging.info(msg)
                        self.missing_rows= self.missing_rows+1
                self.row = []
                self.row.append(word_dot)
            elif len(self.row)==1:
                self.row[0] = self.row[0]+' '+ word_dot
            else:
                self.row.append(word_dot)
            self.prev_complete_name = False
            if len(two_dots)>=1:
                self.prev_complete_name = True
                
            
    def checkIfPartyName(self,name):
        name_corr = re.sub(r'\\n|\[|\]|\'', '',str(name)).strip()
        if name_corr in self.party_set:
            self.previous_party = name_corr.lower().replace('\n', ' ').replace('\r', '')
            return True
        for party_name in self.party_set:

            score = fuzz.ratio(party_name.lower(),name_corr.lower())
            if score>=85:
                self.previous_party = name_corr.lower().replace('\n', ' ').replace('\r', '')
                return True
        if len(name_corr)<=2:
            return False
        Partial_Ratio = fuzz.partial_ratio(name_corr.lower(),"partei")
        if Partial_Ratio>=85:
            self.previous_party = name_corr.lower().replace('\n', ' ').replace('\r', '')
            return True
        return False
    
    def writeToCSV(self):
        logging.info("Approx total rows: "+str(self.total_rows))
        logging.info("Approx accurate rows: "+str(self.total_rows - self.missing_rows))
        logging.info("Approx ignorned rows: "+str(self.missing_rows))
                                        
        self.allRows.insert(0,["Approx total rows: "+str(self.total_rows)])
        self.allRows.insert(1,["Approx accurate rows: "+str(self.total_rows - self.missing_rows)])
        self.allRows.insert(2,["Approx ignorned rows: "+str(self.missing_rows)])
        df = pd.DataFrame(self.allRows)
        df.to_csv("../Data/Output_Step3/"+self.output_filename, sep='\t',index =False)
        
    #This function splits the data into individual words
    def processData(self):
        row_cnt=0
        for eachRow in self.data:
            '''
            if row_cnt==0:
                print(row_cnt," ",eachRow)
                row_cnt=row_cnt+1
                continue
            '''
            #print(row_cnt," ",eachRow)
            row_cnt=row_cnt+1
            for eachElement in eachRow:
                #print(eachElement)
                #check if the value is nan
                if isinstance(eachElement, float) and math.isnan(eachElement):
                    continue
                if self.checkIfPartyName(eachElement):
                    #party_li = [eachElement]
                    #self.allRows.append(party_li)
                    print("partyname: ",eachElement)
                    self.party_cnt=self.party_cnt+1
                    continue
                split_vals = eachElement.split()
                for element in split_vals:
                    self.addStructure(element)
        
        self.writeToCSV()
                #for eachIndi in eachElement:
                #    print(eachIndi)
                #    
                    #for splits in split_vals:
                    #     print("Split :",splits)
                    
    def loadPartyNames(self):        
        for i in range(0,len(self.party_list)):
            val = re.sub(r'\\n|\[|\]|\'', '',str(self.party_list[i])).strip()
            self.party_set.add(val)
            
    
```

```python
from tabula import read_pdf
from tabula import convert_into

class ReadCsvFiles:
    def __init__(self,file_path,file_type="csv"):
        if file_type=="csv":
            self.data_df = pd.read_csv(file_path, header=None, sep='\t')
            self.data_list = self.data_df.values.tolist()
        elif file_type=="pdf":
            self.data_df = read_pdf(file_path, pages="3",multiple_tables=True,area=(400,40,100,100))
    
    def getPdfData(self):
        return self.data_df
    
    def getCsvData(self):
        return self.data_df,self.data_list

def loadMetadata():
        metadata = "../Data/Input_Step3/1930_32/1930_32_Metadata.xlsx"
        meta = pd.ExcelFile(metadata)
        names = meta.parse()['Name'].values[:]
        cnt_votes = meta.parse()['Votes'].values[:]

        name_to_votecnt ={}
        for i in range(0,len(names)):
            name = names[i]
            if names[i][-4:] ==".pdf":
                #print(names[i])
                name =names[i][:-4]
            ch = ""
            for par in name:
                if par=="/":
                    ch = ch+" "
                else:
                    ch = ch+par
                #print(ch)
            #break
            name_to_votecnt[ch] = cnt_votes[i]

        return name_to_votecnt
```

```python

from os import listdir
from os.path import isfile, join

#Initialize all the data
party_data,party_list = ReadCsvFiles("../Data/Input_Step3/party_names.csv").getCsvData()
vote_data,vote_list = ReadCsvFiles("../Data/Input_Step3/vote_names.csv").getCsvData()


mypath = "../Data/Input_Step3/1930_32/Csv_files/"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
onlyfiles.sort()

meta_map = loadMetadata()


files_to_check = []
missing_per = []
files_names = []
party_cnt = []
#onlyfiles= ["1930 32,2.pdf_628_634.csv"]
for file_name in onlyfiles:
    
    #print("Processing.... :",file_name," Expected cols: ",meta_map[file_name[:-4]])
    file_data,file_list = ReadCsvFiles(mypath+file_name).getCsvData()
    #print(file_list)

    #print(party_list)
    #print(vote_list[0][0])
    vote_li = set()
    for i in vote_list:
        for j in i:
            vote_li.add(j.lower())
    print(vote_li)
    #break
    dp = DataProcessor(file_list,file_name.split("\.")[0],party_list,vote_li,exp_cols=meta_map[file_name[:-4]]+1)
    #logging.info("Processing.... :"+file_name+" Expected cols: "+str(meta_map[file_name[:-4]]))
    print("Processing.... :"+file_name+" Expected cols: "+str(meta_map[file_name[:-4]]))
    dp.processData()
    print("party cnt:",dp.party_cnt)
    
    files_names.append(file_name)
    party_cnt.append(dp.party_cnt)
    if dp.missing_rows > 20*(dp.total_rows/100):
        files_to_check.append(file_name)
        missing_per.append(((dp.missing_rows*100)/dp.total_rows))
        
    #break
```
