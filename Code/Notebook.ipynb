```python
import re
import math
def processRow(val):
    new_val = []
    print("Entire array: ",val)
    for i in val:
        print((i))
        
        #check if the value is nan
        if isinstance(i, float) and math.isnan(i):
            #print("Nan value")
            new_val.append("-")
            continue
            
        #replacing all the dots with space
        replace_dot = re.sub('[.]{2,}',"#$",i) #i.replace('.', '#')
        #replace close bracket
        replace_brac = re.sub("\)",")#$",replace_dot)
        #convert multiple whitespaces into single whitespace
        white_rem = replace_brac#re.sub('#+',"#$",replace_dot)
        #print(white_rem+" "+str(len(white_rem)))
        
        new_val.append(white_rem)
    print(new_val," ",str(len(new_val)))
    return new_val
arr_list =[]
for i in range(0,len(li)):        
    arr_list.append(processRow(li[i]))
```
